```python
def get_data(place):
    places = []
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    params = {'input': '{}',
          'inputtype': 'textquery',
          'fields': 'place_id,photos,formatted_address,name,rating,opening_hours,geometry',
          'key': 'AIzaSyAtzvDoPkWFu8g6urdrhaManQq-zccP9RM'
         }
    params['input'] = params['input'].format(place)
    page = requests.get(url, params=params)
    if page.status_code == 200:
        data = page.json()
        if data['status'] == 'OK':
            for each_place in data['candidates']:
                    d = {
                        'place_id': each_place.get('place_id', 'not found'),
                        'address':each_place.get('formatted_address', 'not found'),
                        'name': each_place.get('name', 'not found'),
                        'photos':[
                            ph['photo_reference'] for ph in each_place['photos']
                        ]
                    }
                    places.append(d)
        else:
            print("Error!!", data)
        
    else:
        print("Error!!! Something Went Wrong" , page.status_code, )
    return places
    
```

## Create a tkinter application to search places and show their info and photo in top level window

```python
def get_data(place):
    places = []
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    params = {'input': '{}',
          'inputtype': 'textquery',
          'fields': 'place_id,photos,formatted_address,name,rating,opening_hours,geometry',
          'key': 'AIzaSyAtzvDoPkWFu8g6urdrhaManQq-zccP9RM'
         }
    params['input'] = params['input'].format(place)
    page = requests.get(url, params=params)
    if page.status_code == 200:
        data = page.json()
        if data['status'] == 'OK':
            for each_place in data['candidates']:
                    d = {
                        'place_id': each_place.get('place_id', 'not found'),
                        'address':each_place.get('formatted_address', 'not found'),
                        'name': each_place.get('name', 'not found'),
                        'photos':[
                            ph['photo_reference'] for ph in each_place.get('photos')
                        ]
                    }
                    places.append(d)
        else:
            print("Error!!", data)
        
    else:
        print("Error!!! Something Went Wrong" , page.status_code, )
    return places
    
    
def show_image(photo_reference):
    photo_url = "https://maps.googleapis.com/maps/api/place/photo?maxwidth=640&maxheight=480&photoreference={}&key=AIzaSyAtzvDoPkWFu8g6urdrhaManQq-zccP9RM"
    img_url = photo_url.format(photo_reference)
    page = requests.get(img_url)
    if page.status_code == 200:
        if 'image' in page.headers['Content-type']:
            with open('show_img.jpeg', 'wb') as fp:
                fp.write(page.content)
                fp.close()
            im = plt.imread('show_img.jpeg')
            plt.imshow(im)
            plt.xticks([])
            plt.yticks([])
            plt.show()
        else:
            print("no image found")
    else:
        print("!Error!", page.status_code)
    
```

```python
def search(addrs):
    for address in addrs:
        data = get_data(address)
        if data:
            for each_place in data:
                print("__"*30)
                print('\n\n')
                print("Name: ", each_place['name'])
                print("Address: ", each_place['address'])
                for ph_ref in each_place['photos']:
                    show_image(ph_ref)
        print("\n\n")
        print("__"*60)
        print('\n\n\n')
```

```python
search(['diggi, rajasthan','chittorgarh', 'kanor','neemuch' ])
```

```python
search(['jodhpur'])
```
