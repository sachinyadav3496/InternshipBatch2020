```python
page = requests.get("https://www.rottentomatoes.com/")
```

```python
soup = bs4.BeautifulSoup(page.text)
```

```python
urls = soup.find_all('img', {'slot':'poster'})
```

```python
tomotos = soup.find_all('span', {'slot': 'title'})
len(tomotos)
```

```python
len(urls)
```

```python
urls[:2]
```

```python
def get_urls(tag_list):
    urls = []
    for tag in tag_list:
        url = tag.get('data-src')
        #print(url)
        if url.startswith('http'):
            urls.append(url)
    return urls
```

```python
im_urls = get_urls(urls)
im_urls[-5:]
```

```python
tomotos[0].string
```

```python
tomotos[-5:]
for i in range(len(tomotos)):
    tomotos[i] = tomotos[i].string
```

```python
tomotos[:5]
```

```python
icons = soup.find_all('score-icon-critic')
per = []
for ic in icons:
    try:
        per.append(int(ic.get('percentage')))
    except ValueError:
        per.append(0)
```

```python
len(per)
```

```python
soup.find_all('div',)# slot="critic-score")
```

```python
tomotos
```

```python
df = pd.DataFrame({
    'images': im_urls,
    'titles':  tomotos,
    'popularity':  per
})
```

```python
df.head()
```

```python
finding = df[['titles', 'popularity']].sort_values('popularity', ascending=False)
```
