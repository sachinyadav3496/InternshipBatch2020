```python
import os
def download_images(img_urls, dir_path='images'):
    for num, im_url in enumerate(img_urls):
        img_path = os.path.join(dir_path, f'{num+1}.'+im_url.split('/')[-1])
        page = requests.get(im_url)
        if page.status_code == 200 and 'image' in page.headers['Content-Type']:
            with open(img_path, 'wb') as file:
                file.write(page.content)
                file.close()
            #im = plt.imread(img_path)
            #plt.imshow(im)
            #plt.xticks([]); plt.yticks([])
            #plt.show()
```

```python
def show_images(soup, attrs={},dir_path="./", img_attr='src', show=True):
    images = soup.find_all('img', attrs=attrs)
    print("\nTotal Images Found with given attribute: ", len(images))
    for num, tag in enumerate(images):
        url = tag.get(img_attr)
        if '?' in url:
            i = url.find('?')
            url = url[:i]
        fname = f"{num}."+url.split('/')[-1]
        if url.strip() and url.startswith('http'):
            page = requests.get(url)
        if page.status_code == 200 and 'image' in page.headers['Content-Type']:
            path = os.path.join(dir_path, fname)
            with open(path, 'wb') as fp:
                fp.write(page.content)
                fp.close()
            if show:
                im = plt.imread(path)
                plt.xticks([])
                plt.yticks([])
                plt.imshow(im)
                plt.show()
            
                
            
                
```

```python
show_images(soup, attrs={'class': "lazy img-responsive"}, img_attr='data-original', dir_path='images/test2', 
           show=False)
```
