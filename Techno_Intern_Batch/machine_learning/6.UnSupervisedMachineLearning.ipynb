```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import make_blobs
from  sklearn.cluster import KMeans
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
```

```python
data = make_blobs(n_samples=500, n_features=2, centers=3,
                 cluster_std=2)
```

```python
features, target = data
```

```python
plt.figure(dpi=300)
sns.scatterplot(features[:, 0], features[:, 1], hue=target, palette='rainbow')
plt.show()
```

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.palplot(sns.color_palette(flatui))
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(features)
```

```python
c = model.cluster_centers_
c
```

```python
plt.figure(dpi=300)
sns.scatterplot(features[:, 0], features[:, 1], hue=target, palette='rainbow')
plt.plot(c[0][0], c[0][1], 'kx', ms=10)
plt.plot(c[1][0], c[1][1], 'kx', ms=10)
plt.show()
```

```python
hue = model.labels_
plt.figure(dpi=300)
sns.scatterplot(features[:, 0], features[:, 1], hue=hue, palette='rainbow')
plt.plot(c[0][0], c[0][1], 'kx', ms=10)
plt.plot(c[1][0], c[1][1], 'kx', ms=10)
plt.show()
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(features)
```

```python
model.cluster_centers_
```

```python
c = model.cluster_centers_
plt.figure(dpi=300)
sns.scatterplot(features[:, 0], features[:, 1], hue=target, palette='rainbow')
plt.plot(c[0][0], c[0][1], 'kx', ms=10)
plt.plot(c[1][0], c[1][1], 'kx', ms=10)
plt.plot(c[2][0], c[2][1], 'kx', ms=10)
plt.show()
```

```python
c = model.cluster_centers_
hue = model.labels_
plt.figure(dpi=300)
sns.scatterplot(features[:, 0], features[:, 1], hue=hue, palette='rainbow')
plt.plot(c[0][0], c[0][1], 'kx', ms=10)
plt.plot(c[1][0], c[1][1], 'kx', ms=10)
plt.plot(c[2][0], c[2][1], 'kx', ms=10)
plt.show()
```
