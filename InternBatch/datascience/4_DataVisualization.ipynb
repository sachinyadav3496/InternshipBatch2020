## subplots

```python
fig = plt.figure(figsize=(10,5), edgecolor='red', linewidth=3)

ax1 = fig.add_axes([0, 0, 0.4, 0.4])
ax2 = fig.add_axes([0.6, 0.6, 0.4, 0.4])


ax1.plot(x, x**2, label='square')
ax1.plot(x, x**3, label='cube')
ax1.legend(fontsize=15)

ax2.plot(angle, np.sin(angle), '--r', label='sin')
ax2.plot(angle, np.cos(angle), '--b', label='cos')
ax2.legend(fontsize=10)

sp_labels=['top', 'bottom', 'right', 'left']
axeses = [ax1, ax2]
for axes in axeses:
    for label in sp_labels:
        axes.spines[label].set_visible(False)
    axes.set_xticks([])
    axes.set_yticks([])
#plt.savefig('myfig.png', dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10,5), facecolor='gray')
ax.plot(x, x**3, label='cube')
for sp, color in zip(['top', 'bottom', 'left', 'right'], ['gold', 'silver', 'orange', 'green']):
    ax.spines[sp].set_color(color)
    ax.spines[sp].set_linewidth(5)

ax.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.labelcolor'] = 'green'
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.color'] = 'red'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.color'] = 'red'
plt.rcParams['ytick.labelsize'] = 15
#plots default parameters
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, x**2, '--r')
ax[1].plot(x, x**3, '--g')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, x**2, '--r')
ax2.plot(x, x**3, '--g')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0][0].plot(x, x**2, 'r--')
ax[0][1].plot(x, x**3, 'g--')

ax[1][0].plot(angle, np.sin(angle), 'c--')
ax[1][1].plot(angle, np.cos(angle), 'm--')

plt.show()
```

```python
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

ax1.plot(x, x**2, 'r--')
ax2.plot(x, x**3, 'g--')

ax3.plot(angle, np.sin(angle), 'c--')
ax4.plot(angle, np.cos(angle), 'm--')

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xticks([])
    ax.set_yticks([])
    for label  in ['top', 'bottom', 'right', 'left']:
        ax.spines[label].set_visible(False)

plt.show()
```

```python
fs  = {
    'fontsize': 20,
    'color': '#123456',
    'weight': 'bold',
    'family': 'monospace'
}
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, x**2, label='$x^2$')
ax.plot(x, x**3, label='$x^3$')
ax.set_title('square vs cube', fontdict=fs)
ax.set_xlabel('x values', fontdict=fs)
ax.set_ylabel('y=f(x)', fontdict=fs)

for sp, color in zip(ax.spines, 'cymg'):
    ax.spines[sp].set_color(color)
    ax.spines[sp].set_linewidth(2)
ax.legend(fontsize=20)
ax.grid(color='m', linestyle='--', alpha=0.5)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, x**2, label='$x^2$')
ax.plot(x, x**3, label='$x^3$')
ax.set_title('square vs cube', fontdict=fs)
#ax.set_xlabel('x values', fontdict=fs)
#ax.set_ylabel('y=f(x)', fontdict=fs)

for sp, color in zip(ax.spines, 'cymg'):
    if sp in ['top', 'right']:
        ax.spines[sp].set_visible(False)
    else:
        ax.spines[sp].set_position(('axes', 0.5))
        ax.spines[sp].set_color(color)
        ax.spines[sp].set_linewidth(2)
        ax.spines[sp].set_linestyle('--')
ax.legend(fontsize=20)
ax.grid(color='m', linestyle='--', alpha=0.5)
plt.show()
```
