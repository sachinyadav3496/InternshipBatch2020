```python
sal = pd.DataFrame({
    'sex': [ 'm', 'f', 'm', 'm', 'f', 'f', 'f', 'm'],
    'sal': [ 20, 25, 30, 40, 25, 30, 40, 40]
})
```

```python
sal
```

```python
def group_by(df, col_name, apply_col, func):
    unique = df[col_name].unique()
    d = {}
    for group in unique:
        d[group] = df[apply_col][df[col_name]==group].agg(func)
    return pd.Series(d, name=func.__name__)
    
    
```

```python
ans = group_by(sal, 'sex','sal', np.mean )
```

```python
top_20 = h1b.groupby('EMPLOYER_NAME').agg({
    'PREVAILING_WAGE': np.mean,
    'YEAR': np.size
})
```

```python
top_20.sort_values('PREVAILING_WAGE', ascending=False, inplace=True)
```

```python
top_20 = top_20.rename({"PREVAILING_WAGE":'salary', 'YEAR':'Count'}, axis=1)
```

#### Per Million

```python
top_20['salary'] = top_20['salary'].apply(lambda value: np.round(value / 1000000, 2))
```

```python
top_20.columns
```

```python
top_20.sort_values(by='Count', ascending=False,inplace=True)
```

```python
data = top_20[:30]
```

```python
data.sort_values(by='salary', ascending=False, inplace=True)
```

```python
plt.rcParams['xtick.labelsize'] = 5
```

```python
plt.figure(dpi=300)
sns.barplot(x=data.index, y=data.salary)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(dpi=300)
sns.barplot(x=data.index, y=data.salary,palette='rainbow' )
plt.xticks(rotation=90)
plt.show()
```
