## Data Visulations

```python
import matplotlib.pyplot as plt
```

```python
height = np.random.normal(140, 10, 25)
weight = height*0.5 + np.random.normal(5, 2, 25)
```

```python
fig = plt.figure(figsize=(5, 3), facecolor='#333333', dpi=300)
ax = fig.add_axes([0, 0, 1, 1], facecolor="#123456")
plt.title('Introduction to Graph', color='white', fontsize=20)
plt.xlabel("Height", color='white')
plt.ylabel("Weight", color='white')
ax.scatter(height, weight)
plt.grid(color='white', ls='--')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
