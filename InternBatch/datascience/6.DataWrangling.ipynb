```python
tip = pd.read_csv('seaborn_data/tips.csv')
```

```python
tip.head()
```

```python
tip['sex'].value_counts() # category
```

```python
data = tip['sex'].value_counts() # category
label = list(data.index)
freq = data.values
```

```python
def show_text(color, ax, fs=15,):
    m_h = 0
    heights = [ b.get_height() for b in ax.patches]
    max_height = max(heights)
    pad = max_height*0.3
    text_pad = max_height*0.1
    for c,box in zip(color, ax.patches):
        x = box.get_x()
        x += box.get_width()/2
        y = box.get_height()
        ax.text(x, y+text_pad, box.get_height(), 
                 color=c, fontsize=fs)
        box.set_color(c)
    
    ax.set_ylim([0, max_height+pad])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
```

```python
fig, ax = plt.subplots()
ax.bar(label, freq)
ax.set_xlabel('Sex')
ax.set_ylabel('Frequency')
show_text(['cyan', 'gold'], ax)

plt.show()
```

#### Stack Bar Chart

```python
fig, ax = plt.subplots()
ax.bar(label, freq, color=['cyan', 'gold'])
ax.bar(label, freq1, color=[ 'purple', 'orange'], bottom=freq)
ax.set_xlabel('Sex')
ax.set_ylabel('Frequency')
ax.set_yticks([])
for sp in ['top', 'bottom', 'left', 'right']:
    ax.spines[sp].set_visible(False)
plt.show()
# put frequencies as text with white color inside each bar # assignment
```

```python
fig, ax = plt.subplots()
ax.set_title("Group Bar Chart", fontsize=20)
ax.bar([0-0.16, 1-0.16], freq, width=0.35, color=['cyan', 'gold'], )
ax.bar([0+0.16,1+0.16], freq1, width=0.35, color=['purple', 'orange'])

ax.set_xticks([0, 1])
ax.set_xticklabels(label)
plt.show()
```
