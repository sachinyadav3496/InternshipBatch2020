### Normalization

    A and B column have different magnitue than it's hard to compare them also if we apply some 
    mathematical operations they will be highly baised on magnitue of those columns

    the value of each column should be in same range such that -1, 1

```python
tip['tip'].max()
```

```python
tip['tip'].min()
```

```python
tip['tip'][127]
```

```python
n = (tip['tip'][127] - tip['tip'].min()) / tip['tip'].max()
```

```python
n
```

```python
tips = tip[['total_bill', 'tip']].copy()
```

```python
tips['tip'].apply(lambda value: value**2, )
```

```python
mi = tips['tip'].min()
mx = tips['tip'].max()

tips['tip_normal'] = tips['tip'].apply(lambda value: (value-mi)/mx)
```

```python
tips.head()
```

```python
mi = tips['total_bill'].min()
mx = tips['total_bill'].max()

tips['total_bill_normal'] = tips['total_bill'].apply(lambda value: (value-mi)/mx)
```

```python
tips.head()
```

```python
tips.hist('total_bill',bins=30, ec='r', color='gold')
plt.show()
```

```python
tips.hist('total_bill_normal', bins=30, ec='r', color='gold')
plt.show()
```

```python
tips[['total_bill', 'total_bill_normal',]].hist( bins=30, ec='r')
plt.show()
```

```python

tips[['tip', 'tip_normal',]].hist( bins=30, ec='r')
plt.show()
```

```python
tips[['total_bill_normal', 'tip_normal']].plot(kind='hist', bins=30,
                                              histtype='step')
plt.show()
```
